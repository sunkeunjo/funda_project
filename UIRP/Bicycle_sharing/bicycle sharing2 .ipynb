```python
import pandas as pd 
from datetime import datetime 

dir_="C://Users/sunkeun_jo/Desktop/data/etl_ymd.csv"
```

```python
etl_data=pd.read_csv(dir_)

etl_data['etl_ymd_Type']=etl_data['etl_ymd'].apply(lambda x : str(x)[0:4]+"-"+str(x)[4:6]+"-"+str(x)[6:])
etl_data['etl_ymd_dateType']=etl_data['etl_ymd_Type'].apply(lambda x: datetime.strptime(x,"%Y-%m-%d"))
etl_data.head()
```

```python
temp=pd.read_csv("C://Users/sunkeun_jo/Desktop/data/ulsan_temp.csv",encoding="utf-8")
temp.head()
```

```python
temp.columns=["시간","기온","강수량","풍속","습도","적설"]
temp.head()
```

```python
etl_data.head()
```

temp 데이터에서 일별 최고 기온 최저 기온으로 데이터를 합쳐달라고 하셨습니다

```python
print(type(temp.loc[0,"시간"])) 
```

우선 시간 type을 datetime으로 바꿔주도록 합시다

```python
temp['시간']=pd.to_datetime(temp['시간'])
print(type(temp.loc[0,"시간"]))
```

```python
#str으로 바꿔줍니다
temp['renewed']=temp['시간'].apply(lambda x: str(x.year)+"-"+str(x.month)+"-"+str(x.day))
temp.head()
```

```python
#다시 우리가 했던 방식대로 바꿔줍니다.
temp['final_renewed']=temp['renewed'].apply(lambda x:datetime.strptime(x,"%Y-%m-%d") )
temp.head(20)
```

```python
#final 데이터를 기준으로 묶어줍니다
new_temp=temp.groupby("final_renewed")
new_temp.get_group("2018-01-01")
```

```python
#큰 것은 큰 값에 , 작은 것은 작은 값에 value들을 넣어줍니다
max_v= new_temp["기온"].max().values
min_v=new_temp["기온"].min().values
```

```python
#date는 순서대로 놓아줍니다
date=temp['final_renewed'].unique()
```

```python
#df 만들기 완성! 
df=pd.DataFrame()
df["etl_ymd_dateType"]=date
df['max_temp']=max_v
df['min_temp']=min_v
df.head()
```

```python
etl_data=pd.read_csv(dir_)
etl_data['etl_ymd_Type']=etl_data['etl_ymd'].apply(lambda x : str(x)[0:4]+"-"+str(x)[4:6]+"-"+str(x)[6:])
etl_data['etl_ymd_dateType']=etl_data['etl_ymd_Type'].apply(lambda x: datetime.strptime(x,"%Y-%m-%d"))
etl_data.head()
```

이제 조교님이 주신 데이터와 병합을 해보겠습니다.

병합에는 `pd.merge(데이터프레임,데이터프레임)`을 이용하도록 하겠습니다

```python
final_df=pd.merge(etl_data,df,on="etl_ymd_dateType")
# 특정열을 기준으로 병합하겠습니다. 
final_df.head()
```

```python
final_df.drop(['etl_ymd_Type','etl_ymd'],axis=1,inplace=True)
final_df
```

```python
#이제 저장해줍시다
final_df.to_csv("final_temp.csv")
```
