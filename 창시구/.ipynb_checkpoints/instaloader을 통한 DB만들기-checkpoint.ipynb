그렇다면 이제 hilton호텔의 피드로 가서 필요한 정보들을 뽑아와 봅시다

```python
#해당 profile 
profile= instaloader.Profile.from_username(L.context,"hilton")
post_list=[]

# hilton의 모든 post를 한 번 뽑아보도록 하겠습니다
print("fetching all posts of a profile {}".format(profile.username))

#우선 10개만 
for post in profile.get_posts():
    post_list.append(post)
    if len(post_list)==10:
        break
print(post_list)
```

post.comments : post의 comment 갯수를 알려준다 

post.date : post의 date를 뽑아서 올려줍니다

post.profile: post를 한 주체를 뽑아서 올려줍니다 

post.typename : post의 type을 뽑아줍니다.

post.captionhastags : post caption에서 어떤 hashtag가 써졌는지를 뽑아 줍니다. #hilton은 caption 란에다가는 쓰지 않았군요;;;;

post.caption :post의 caption에 어떤 글을 썼는지 뽑아서 줍니다

post.caption_mentions: post에서 mention을 했는지 안했는지를 보여줍니다.

post.is_video : video type일 경우 True를 뽑아줍니다 

post.video_view_count : video 시청 수를 보여줍니다

post.video_duration : video 의 영상 길이 

post.likes: post의 like개수를 보여줍니다 

post.get_comments() : 모든 comment를 얻습니다 iteration 형태로 얻기 때문에 이 안에서 특정 정보를 뽑아내서 추가시켜야할 것 같습니다

post.get_likes() : 좋아요를 누른 user들을 보여줍니다 하지만 이 역시 iterator이기 때문에 추가시키면서 넣어야할 것 같습니다 

getcomments()를통해서  answers를 돌리면서 가지고 있는 모든 answer들을 뽑을 수 있습니다. profile을 제외하고 우선 데이터를 뽑아보도록 합시다! 


```python
import pandas as pd 
import tqdm

i=1
content_list=[]
for post in tqdm.tqdm(profile.get_posts()):
    post_list.append(post)
    com_list=[]
    for com in post.get_comments():
        ans_list=[]
        for ans in com.answers:
            ans_list.append([ans.text,ans.owner,ans.likes_count])
        com_list.append([com.text,com.likes_count,com.owner,ans_list])
    content_list.append([i,post.date,post.profile,post.typename,post.caption_hashtags,post.is_video,post.video_view_count,post.video_duration,post.likes,post.comments,com_list])
    
    i+=1
    
df=pd.DataFrame(content_list,columns=["id","date","profile","type","caption_and_hashtags","is_video","view_count","video_duration","likes","comments","com_list"])
df.head()
df.to_csv("hilton_post2",encoding="utf-8")
```
