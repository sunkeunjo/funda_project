```python
# 데이터 불러오기 
import pandas as pd 
# 경고 무시하기
import warnings 
warnings.filterwarnings("ignore",category=DeprecationWarning)
import re
post=pd.read_csv("final_DB")
comment=pd.read_csv("final_comment_DB")
post.head()
# comment.head(3)
```

```python
post["profile"].unique()
```

```python
# 우선 caption에 대한 cleaning 시작 
def clean_text(text):
    # 모두 소문자로 바꿔줍니다,@+word는 모조리 제거해 줍니다.글자 안에 숫자가 들어간 것들도 바꿔줍니다.
    
    text=str(text).lower()
    text=re.sub("@.*","",text)
    text=re.sub("\w*\d\w*","",text)
    text=re.sub("#.*","",text)
    
    return text
```

```python
post["caption"]=post["caption"].apply(lambda x: clean_text(x))
post.head(3)
```

lets remove stop words and count the common word 

```python
import nltk
from nltk.corpus import stopwords

SW=set(stopwords.words("english"))

print(len(SW))
```

```python
def tokenize_text(text):
    words_list=text.split()
    
    text= [word for word in words_list if not word in SW if len(word)>1]
    
    return text
```

```python
post["caption_token"]=post["caption"].apply(lambda x: tokenize_text(x))
post.head()
```

```python
%%time
# sentiment 다시 
from nltk.sentiment.vader import SentimentIntensityAnalyzer

def compound_score(text):
    senti_analyzer=SentimentIntensityAnalyzer()
    sentence=" ".join(text)
    senti_scores=senti_analyzer.polarity_scores(sentence)
    return senti_scores["compound"]

def positive_score(text):
    senti_analyzer=SentimentIntensityAnalyzer()
    sentence=" ".join(text)
    senti_scores=senti_analyzer.polarity_scores(sentence)
    return senti_scores["pos"]

def neutral_score(text):
    senti_analyzer=SentimentIntensityAnalyzer()
    sentence=" ".join(text)
    senti_scores=senti_analyzer.polarity_scores(sentence)
    return senti_scores["neu"]

def negative_score(text):
    senti_analyzer=SentimentIntensityAnalyzer()
    sentence=" ".join(text)
    senti_scores=senti_analyzer.polarity_scores(sentence)
    return senti_scores["neg"]

post["compound"]=post["caption_token"].apply(lambda x: compound_score(x))
post["positive"]=post["caption_token"].apply(lambda x: positive_score(x))
post["negative"]=post["caption_token"].apply(lambda x: negative_score(x))
post["neutral"]=post["caption_token"].apply(lambda x: neutral_score(x))
post.head()
```

```python
post.to_csv("final_DB",index=False)
```
