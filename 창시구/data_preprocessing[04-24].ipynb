```python
### comment ÎΩëÏïÑÎÇ¥Í∏∞ 
def com_list(list_):
    return [list_[x][0] for x in range(len(list_))]

def like_list(list_):
    return [list_[x][1] for x in range(len(list_))]

def comment_writer_list(list_):
    return [comment_profile(list_[x][2]) for x in range(len(list_))]

def comment_comment(list_):
    return [list_[x][3] for x in range(len(list_))]
#######test###########
# new_df['com_list'][0]
# print(com_list(hilton['com_list'][0]))
# print(like_list(hilton['com_list'][0]))
# print(comment_writer_list(hilton["com_list"][3]))
hilton["com_list"][0][0][3]
```

```python
post_df=pd.DataFrame(columns=["post_id","comment","comment_like","comment_writer","Recomment"])
post_df
# hilton
post_df["post_id"]=hilton["id"].values
post_df["comment"]=hilton['com_list'].apply(lambda x: com_list(x))
post_df['comment_like']=hilton["com_list"].apply(lambda x: like_list(x))
post_df["comment_writer"]=hilton["com_list"].apply(lambda x: comment_writer_list(x))
post_df["Recomment"]=hilton["com_list"].apply(lambda x : comment_comment(x))
post_df.to_csv("hilton_comment",index=False)
                                                
```
