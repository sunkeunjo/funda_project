```python
import re 

def contain_list(sentence):
    _list=[]
    pattern= re.compile("^[@].+")#@Í∞Ä Ïò§Í≥† Îí§Ïóê Ïñ¥Îñ§ Îã®Ïñ¥Í∞Ä ÌïòÎÇò Ïù¥ÏÉÅÏôÄÏïºÌï©ÎãàÎã§ 
    # photo byÎùºÎäî Îã®Ïñ¥ÎåÄÏã† @ Í∞Ä ÏûàÎäîÍ≤ΩÏö∞ÏóêÎßå user generated contentÎùºÍ≥† Ï†ïÏùòÌï©ÎãàÎã§. 
    
    for word in sentence.split(" "):
        try :
            pattern.match(word).group(0)
            _list.append(pattern.match(word).group(0))
        except: 
            continue
        
    return _list

def real_ugc(sentence):
    # Î≥¥ÌÜµ ugcÏóêÏÑúÎäî Ïù¥Îü∞ ÌëúÏãúÍ∞Ä Ï°¥Ïû¨ÌïúÎã§. 
    try:
        re.compile("üì∏|photo by").search(sentence.lower()).group(0) 
        ugc=True
    except: 
        ugc=False
    # Î™ÖÏãúÏ†ÅÏúºÎ°ú photo byÎùºÍ≥† Ï†ÅÏùÄ Î∂ÄÎ∂ÑÎèÑ Ï†úÏô∏ 

    return ugc
```

```python
result=wyndham['caption'][:10].apply(lambda x :real_ugc(x))
result
```

```python
for hotel  in [hilton,hyatt,wyndham,holiday]:
    hotel["caption_tag"]=hotel['caption'].apply(lambda x: contain_list(str(x)))
    hotel['real_ugc']=hotel['caption'].apply(lambda x : real_ugc(x))
    
hilton.head()
```
